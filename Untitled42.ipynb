```python
from spire.xls import *
from spire.xls.common import *

# Create a Workbook object
workbook = Workbook()

# Load an Excel document
workbook.LoadFromFile("C:/Users/Shreshtha/Downloads/Bartrack Sample.xlsx")

# Iterate through the worksheets in the workbook
for sheet in workbook.Worksheets:
    # Get the PageSetup object
    pageSetup = sheet.PageSetup

    # Set page margins
    pageSetup.TopMargin = 0.1
    pageSetup.BottomMargin = 0.1
    pageSetup.LeftMargin = 0.1
    pageSetup.RightMargin = 0.1

    # Set worksheet to fit to page when converting
    workbook.ConverterSetting.SheetFitToPage = True

    # Add sheet title as a text box outside the right margin
    text_box = sheet.TextBoxes.AddTextBox(-1, -1, 20, 100)  # Adjusted X-coordinate
    text_box.Text = sheet.Name

# Convert to PDF file
workbook.SaveToFile("C:/Users/Shreshtha/Downloads/Bartrack-Sample-pdf29.pdf", FileFormat.PDF)

# Dispose of the workbook
workbook.Dispose()
```

```python
from spire.xls import *
from spire.xls.common import *

# Create a Workbook object
workbook = Workbook()

# Load an Excel document
workbook.LoadFromFile("C:/Users/Shreshtha/Downloads/Bartrack Sample.xlsx")

# Iterate through the worksheets in the workbook
for sheet in workbook.Worksheets:
    # Get the PageSetup object
    pageSetup = sheet.PageSetup

    # Set page margins
    pageSetup.TopMargin = 0.1
    pageSetup.BottomMargin = 0.1
    pageSetup.LeftMargin = 0.1
    pageSetup.RightMargin = 0.1

    # Set worksheet to fit to page when converting
    workbook.ConverterSetting.SheetFitToPage = True

    # Add sheet title as a text box outside the right margin
    sheet_name = sheet.Name
    text_box_width = len(sheet_name) + 10  # Adjust the factor as needed
    left_margin = pageSetup.LeftMargin * 36  # Convert inches to points
    top_margin = pageSetup.TopMargin * 36  # Convert inches to points
    text_box = sheet.TextBoxes.AddTextBox(1, 1, text_box_width, 100)  # Adjust the position as needed
    text_box.Text = sheet_name

    # Add additional text box for header content
    header_text = "Header: " + sheet_name
    header_text_box = sheet.TextBoxes.AddTextBox(1, 1, 200, 20)  # Adjust the position and size as needed
    header_text_box.Text = header_text
    

# Convert to PDF file
workbook.SaveToFile("C:/Users/Shreshtha/Downloads/Bartrack-Sample-pdf43.pdf", FileFormat.PDF)

# Dispose of the workbook
workbook.Dispose()
```
