```python
import shutil
import pathlib
import os

def extract_images_from_excel(path, output_folder_name='extracted_images'):
    if not isinstance(path, pathlib.Path):
        path = pathlib.Path(path)

    name = path.stem
    output_folder = path.parent / output_folder_name
    output_folder.mkdir(exist_ok=True)

    media_dir = path.parent / 'xl' / 'media'

    new_paths = []

    for i, file in enumerate(os.listdir(media_dir)):
        image_path = media_dir / file
        new_path = output_folder / f'{name}-{i + 1}.png'
        shutil.copy(image_path, new_path)
        new_paths.append(new_path)

    return new_paths
```
