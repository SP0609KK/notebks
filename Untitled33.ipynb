```python
import win32com.client

def identify_object_type(excel_file_path, sheet_name='Sheet6'):
    try:
        excel_app = win32com.client.Dispatch("Excel.Application")
        excel_app.Visible = True  # You can set this to False if you don't want Excel to be visible

        workbook = excel_app.Workbooks.Open(excel_file_path)
        sheet = workbook.Sheets(sheet_name)

        for shape in sheet.Shapes:
            print(f"Object Type: {shape.Type}")

    except Exception as e:
        print(f"Error: {e}")

    finally:
        workbook.Close(False)
        excel_app.Quit()

# Replace 'your_excel_file.xlsx' with the actual path to your Excel file
excel_file_path = 'C:/Users/Shreshtha/Downloads/Sample MNS Bartrack Report1.xlsx'
sheet_name = 'Sheet6'

identify_object_type(excel_file_path, sheet_name)
```
