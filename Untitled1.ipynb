```python
class Rectangle:
    def __init__(self, length, breadth):
        self.length = length
        self.breadth = breadth
    def get_area(self):
        return self.length * self.breadth
    
```

```python
obj =Rectangle(100,120)
```

```python
obj.get_area()
```

```python
class Student:
    def __init__(self, age =0):
        self._age = age
    def get_age(self):
        return self._age
    def set_age(self, x):
        self._age = x
```

```python
obj1 = Student()
```

```python
obj1.get_age()
```

```python
obj1.set_age(30)
```

```python
obj1.get_age()
```
