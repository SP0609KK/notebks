```python
import fitz  # PyMuPDF
from PIL import Image
import pytesseract

# Set the path to the Tesseract executable (change this according to your installation)
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

def extract_text_from_pdf(pdf_path):
    text = ""
    pdf_document = fitz.open(pdf_path)

    for page_num in range(pdf_document.page_count):
        # Convert each PDF page to an image
        page = pdf_document[page_num]
        image = page.get_pixmap()
        image_path = f'tmp_image_{page_num}.png'

        # Save the Pixmap as an image using PIL
        image_pil = Image.frombytes("RGB", [image.width, image.height], image.samples)
        image_pil.save(image_path)

        # Extract text from the image using pytesseract
        image_text = extract_text_from_image(image_path)
        text += image_text

    pdf_document.close()
    return text

def extract_text_from_image(image_path):
    # Preprocess the image if needed (e.g., resize, enhance contrast)
    # You may need to experiment with different preprocessing steps
    image = Image.open(image_path)

    # Extract text from the image using pytesseract
    text = pytesseract.image_to_string(image)

    return text

# Example usage
pdf_path = "C:/Users/Shreshtha/Downloads/Sample MNS Bartrack Report try pdf.pdf"
pdf_text = extract_text_from_pdf(pdf_path)
print(pdf_text)
```
