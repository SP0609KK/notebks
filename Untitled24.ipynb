```python
import json

# Specify the path to your JSON file
json_file_path = "C:/Users/Shreshtha/Downloads/file.txt"

# Load the JSON file into a Python dictionary
with open(json_file_path, "r") as json_file:
    extracted_data = json.load(json_file)

# Function to format JSON-like structure for human readability
def format_json_structure(data, indent=0):
    result = ""
    for key, value in data.items():
        if isinstance(value, dict):
            result += f"{' ' * indent}{key}:\n{format_json_structure(value, indent + 2)}"
        else:
            result += f"{' ' * indent}{key}: {value}\n"
    return result

# Format the extracted data for human readability
formatted_text = format_json_structure(extracted_data)

# Print the formatted text directly in Python
print(formatted_text)
```

```python
import json
import pandas as pd
from pandas import json_normalize

# Specify the path to your JSON file
json_file_path = "C:/Users/Shreshtha/Downloads/file.txt"

# Load the JSON file into a Python dictionary
with open(json_file_path, "r") as json_file:
    extracted_data = json.load(json_file)

# Flatten the nested JSON structure using json_normalize
flattened_data = json_normalize(extracted_data)

# Filter columns with simple string values
flattened_data = flattened_data.select_dtypes(include=['object'])

# Print the resulting DataFrame
print(flattened_data)
```
