```python
import pandas as pd
from google.cloud import documentai_v1 as documentai

def extract_text_from_image(project_id: str, location: str, processor_id: str, image_path: str, mime_type: str) -> str:
    """
    Extracts text from an image using Document AI.
    """

    opts = {"api_endpoint": f"{location}-documentai.googleapis.com"}
    documentai_client = documentai.DocumentUnderstandingServiceClient(client_options=opts)

    # The full resource name of the processor
    resource_name = documentai_client.processor_path(project_id, location, processor_id)

    # Use Document AI Form Parser to extract text from the image
    image = documentai.types.Image(content=image_path)  # No need for gs:// prefix
    raw_document = documentai.types.RawDocument(content=image)

    # Configure the process request
    request = documentai.types.ProcessRequest(
        name=resource_name, raw_document=raw_document
    )

    # Use the Document AI client to process the image
    result = documentai_client.process(request=request)

    # Extracted text from the result
    extracted_text = ""
    for page in result.document.pages:
        for block in page.blocks:
            extracted_text += block.text.content + "\n"

    return extracted_text.strip()

# Replace these with your actual values
PROJECT_ID = "orbital-lantern-412715"
LOCATION = "us"  # Format is 'us' or 'eu'
PROCESSOR_ID = "c770054d62c21621"  # Create processor in Cloud Console
IMAGE_PATH = "trialbuckk1/Trial 2-2.png"  # Replace with your GCS image file path
MIME_TYPE = "image/png"  # Specify the MIME type for your image

# Extract text from the image using the modified function
extracted_text = extract_text_from_image(
    project_id=PROJECT_ID,
    location=LOCATION,
    processor_id=PROCESSOR_ID,
    image_path=IMAGE_PATH,
    mime_type=MIME_TYPE,
)

# Print the extracted text
print(extracted_text)
```
