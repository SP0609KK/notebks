```python
import cv2
import pytesseract

# Read the image using OpenCV
image = cv2.imread("C:/Users/Shreshtha/Downloads/Project UHC/extracted_images/Trial 2-2.png")

# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Use adaptive thresholding for improved text visibility
_, thresh_image = cv2.threshold(gray_image, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)

# Use PyTesseract for OCR with bounding boxes
custom_config = r'--oem 3 --psm 6 outputbase digits'
boxes = pytesseract.image_to_boxes(thresh_image, config=custom_config)

# Extract text and coordinates
extracted_info = []
for box in boxes.splitlines():
    b = box.split()
    text = b[0]
    x, y, w, h = int(b[1]), int(b[2]), int(b[3]), int(b[4])
    extracted_info.append({'text': text, 'x': x, 'y': y, 'w': w, 'h': h})

# Sort the extracted information based on y-coordinate for each line
extracted_info.sort(key=lambda x: x['y'])

# Draw bounding boxes on the original image
for item in extracted_info:
    x, y, w, h = item['x'], item['y'], item['w'], item['h']
    cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)

# Display the image with bounding boxes
cv2.imshow('Extracted Text', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
