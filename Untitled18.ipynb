```python
import cv2
import pytesseract

# Read the image using OpenCV
image = cv2.imread("C:/Users/Shreshtha/Downloads/Project UHC/extracted_images/Trial 2-2.png")

# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Use adaptive thresholding for improved text visibility
_, thresh_image = cv2.threshold(gray_image, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)

# Use PyTesseract for OCR
custom_config = r'--oem 3 --psm 6 outputbase digits'
extracted_text = pytesseract.image_to_string(thresh_image, config=custom_config)

# Print the extracted text
print("Extracted Text:")
print(extracted_text)
```

```python
import cv2
import pytesseract

# Set the path to the Tesseract executable (change this path to the location on your machine)
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

# Read the image using OpenCV
image_path = "C:/Users/Shreshtha/Downloads/Project UHC/extracted_images/Trial 2-2.png"
image = cv2.imread(image_path)

# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Use adaptive thresholding for improved text visibility
_, thresh_image = cv2.threshold(gray_image, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)

# Use PyTesseract for OCR
text = pytesseract.image_to_string(thresh_image)

# Organize text to resemble form structure
lines = text.split('\n')
formatted_text = ''
for line in lines:
    # Attempt to split lines based on various delimiters
    for delimiter in [':', '=', '-', '|']:
        if delimiter in line:
            key, value = map(str.strip, line.split(delimiter, 1))
            formatted_text += f"{key}: {value}\n"
            break  # Break after the first successful split

# Print or use the formatted text
print(formatted_text)
```

```python
import cv2
import pytesseract
import spacy

# Load spaCy English model
nlp = spacy.load("en_core_web_sm")

# Set the path to the Tesseract executable (change this path to the location on your machine)
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

# Read the image using OpenCV
image_path = "C:/Users/Shreshtha/Downloads/Project UHC/extracted_images/Trial 2-2.png"
image = cv2.imread(image_path)

# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Use adaptive thresholding for improved text visibility
_, thresh_image = cv2.threshold(gray_image, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)

# Use PyTesseract for OCR
text = pytesseract.image_to_string(thresh_image)

# Process the text with spaCy
doc = nlp(text)

# Extract named entities
named_entities = [(ent.text, ent.label_) for ent in doc.ents]

# Print or use the extracted named entities
for entity, label in named_entities:
    print(f"{label}: {entity}")
```
