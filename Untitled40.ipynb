```python
import PyPDF2

def merge_pdf_pages(header_page, content_page):
    merged_page = PyPDF2.pdf.PageObject.createBlankPage(width=595, height=841)  # Assuming A4 size

    # Merge the header content
    merged_page.mergeTranslatedPage(header_page, 0, 0)

    # Merge the content below the header
    merged_page.mergeTranslatedPage(content_page, 0, -40)  # Adjust the offset based on your content position

    return merged_page

def merge_pdfs(header_pdf_path, content_pdf_path, output_pdf_path):
    # Open the PDFs
    with open(header_pdf_path, 'rb') as header_pdf_file, open(content_pdf_path, 'rb') as content_pdf_file:
        header_pdf_reader = PyPDF2.PdfFileReader(header_pdf_file)
        content_pdf_reader = PyPDF2.PdfFileReader(content_pdf_file)

        # Create a PDF writer for the merged output
        pdf_writer = PyPDF2.PdfFileWriter()

        # Get the number of pages in the smaller of the two PDFs
        num_pages = min(header_pdf_reader.numPages, content_pdf_reader.numPages)

        # Merge each corresponding page from header and content PDFs
        for i in range(num_pages):
            header_page = header_pdf_reader.getPage(i)
            content_page = content_pdf_reader.getPage(i)

            merged_page = merge_pdf_pages(header_page, content_page)

            # Add the merged page to the output PDF
            pdf_writer.addPage(merged_page)

        # Write the merged PDF to the output file
        with open(output_pdf_path, 'wb') as output_pdf_file:
            pdf_writer.write(output_pdf_file)

# Example usage
header_pdf_path = "C:/Users/Shreshtha/Downloads/Bartrack-Sample-pdf2.pdf"
content_pdf_path = "C:/Users/Shreshtha/Downloads/Bartrack-Sample-pdf1.pdf"
output_pdf_path = "C:/Users/Shreshtha/Downloads/Bartrack-Sample-pdf-merged.pdf"
merge_pdfs(header_pdf_path, content_pdf_path, output_pdf_path)
```
