```python
import pdfplumber
import fitz  # PyMuPDF

def extract_sheet_names(sheet_names_pdf):
    with pdfplumber.open(sheet_names_pdf) as pdf:
        sheet_names = [page.extract_text().strip() for page in pdf.pages]
    return sheet_names

def add_sheet_names_to_pdf(sheet_names, content_pdf, output_pdf):
    input_pdf = fitz.open(content_pdf)
    for i, sheet_name in enumerate(sheet_names):
        page = input_pdf[i]
        page.insert_text((10, 10), sheet_name)  # Adjust the position as needed
    input_pdf.save(output_pdf)

# Example usage
sheet_names_pdf = "C:/Users/Shreshtha/Downloads/Revised_MNS_Sample_Headers_only.pdf"
content_pdf = "C:/Users/Shreshtha/Downloads/Revised_MNS_Sample.pdf" 

sheet_names = extract_sheet_names(sheet_names_pdf)
add_sheet_names_to_pdf(sheet_names, content_pdf, 'Labelled_MNS_Sample.pdf')
```
