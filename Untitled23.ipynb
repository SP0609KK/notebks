```python
from google.cloud import documentai_v1beta3 as documentai

def process_document(project_id='orbital-lantern-412715', input_uri='gs://trialbuckk1/Trial 2-2.png'):
    client = documentai.DocumentUnderstandingServiceClient()

    # Specify the document content and language
    gcs_source = documentai.types.GcsSource(uri=input_uri)
    input_config = documentai.types.InputConfig(
        gcs_source=gcs_source, mime_type='Trial 2-2/png'
    )

    # Specify the feature to extract (e.g., DOCUMENT_TEXT, ENTITY_EXTRACTION, etc.)
    feature = documentai.types.DocumentProcessorType.DOCUMENT_TEXT
    document_processor = documentai.types.DocumentProcessor(
        name=f'projects/{project_id}/locations/us/processors/a61685e7daf98eb7',
        raw_document=documentai.types.RawDocument(content=input_config),
        document_type=feature,
    )

    # Perform document processing
    request = documentai.types.ProcessRequest(
        name=f'projects/{project_id}/locations/us/processors/a61685e7daf98eb7',
        raw_document=documentai.types.RawDocument(content=input_config),
    )

    result = client.process_document(request=request)

    # Extract information from the processed document
    for entity in result.entities:
        print(f'Entity type: {entity.type_}, Mention text: {entity.mention_text}')

if __name__ == "__main__":
    process_document()
```

```python
from google.cloud import documentai_v1beta3 as documentai

def process_image(project_id='orbital-lantern-412715', input_uri='gs://trialbuckk1/Trial 2-2.png'):
    client = documentai.DocumentUnderstandingServiceClient()

    # Specify the document content and language
    gcs_source = documentai.types.GcsSource(uri=input_uri)
    input_config = documentai.types.InputConfig(
        gcs_source=gcs_source, mime_type='image/png'  # Adjust mime_type based on your image format
    )

    # Specify the feature to extract (PROCESS_IMAGE for OCR on images)
    feature = documentai.types.DocumentProcessorType.IMAGE
    document_processor = documentai.types.DocumentProcessor(
        name=f'projects/{project_id}/locations/us/processors/a61685e7daf98eb7',
        raw_document=documentai.types.RawDocument(content=input_config),
        document_type=feature,
    )

    # Perform document processing
    request = documentai.types.ProcessRequest(
        name=f'projects/{project_id}/locations/us/processors/a61685e7daf98eb7',
        raw_document=documentai.types.RawDocument(content=input_config),
    )

    result = client.process_document(request=request)

    # Extract information from the processed document
    for entity in result.entities:
        print(f'Entity type: {entity.type_}, Mention text: {entity.mention_text}')

if __name__ == "__main__":
    process_image()
```

```python
import json

# Specify the path to your JSON file
json_file_path = "C:/Users/Shreshtha/Downloads/file.txt"

# Load the JSON file into a Python dictionary
with open(json_file_path, "r") as json_file:
    extracted_data = json.load(json_file)

# Function to format JSON-like structure for human readability
def format_json_structure(data, indent=0):
    result = ""
    for key, value in data.items():
        if isinstance(value, dict):
            result += f"{' ' * indent}{key}:\n{format_json_structure(value, indent + 2)}"
        else:
            result += f"{' ' * indent}{key}: {value}\n"
    return result

# Format the extracted data for human readability
formatted_text = format_json_structure(extracted_data)

# Print the formatted text directly in Python
print(formatted_text)
```
