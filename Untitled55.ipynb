```python
import tabula

# Provide the path to your PDF file
pdf_path = "C:/Users/Shreshtha/Labelled_MNS_Sample.pdf"

# Extract tables from all pages of the PDF
tables = tabula.read_pdf(pdf_path)

# Print the extracted tables
for page_num, page_tables in enumerate(tables, start=1):
    print(f"Tables from page {page_num}:")
    if page_tables is None:
        print("No tables found on this page.")
    else:
        for table_num, table in enumerate(page_tables, start=1):
            print(f"Table {table_num}:")
            print(table)
```
