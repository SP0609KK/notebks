```python
from openpyxl import load_workbook

def sheets_with_images(bartrack_path):
    try:
        # Load Bartrack workbook
        workbook = load_workbook(bartrack_path, read_only=True)

        sheets_with_images = []

        # Access each sheet and check for images
        for sheet_name in workbook.sheetnames:
            sheet = workbook[sheet_name]

            if hasattr(sheet, '_images') and sheet._images:
                sheets_with_images.append(sheet_name)

        # Close workbook
        workbook.close()

        return sheets_with_images

    except Exception as e:
        print(f"Error checking sheets with images: {e}")
        return None

# Replace 'path/to/Bartrack.xlsx' with the actual path to your Bartrack Excel workbook
bartrack_path = "C:/Users/Shreshtha/Downloads/Copy of Book1 (1).xlsx"
sheets_with_images_result = sheets_with_images(bartrack_path)

if sheets_with_images_result:
    print("Sheets with images:", sheets_with_images_result)
else:
    print("No sheets with images found.")
```

```python
import os
from openpyxl import load_workbook
import pyautogui

def extract_pictures_from_excel(excel_path, output_folder):
    workbook = load_workbook(excel_path)

    for sheet_name in workbook.sheetnames:
        sheet = workbook[sheet_name]

        for image in sheet._images:
            # Assuming that the image is visible on the screen
            left, top, right, bottom = image.anchor.to_coords()
            screenshot = pyautogui.screenshot(region=(left, top, right - left, bottom - top))

            image_path = os.path.join(output_folder, f"{sheet_name}_image_{image.index}.png")
            screenshot.save(image_path)

            print(f"Image saved: {image_path}")

# Replace 'path/to/your/excel/file.xlsx' and 'path/to/your/output/folder' with actual paths
input_excel_path = "C:/Users/Shreshtha/Downloads/Copy of Book1 (1).xlsx"
output_folder_path = "C:/Users/Shreshtha/Downloads/hcu-main/hcu-main/"

# Create the output folder if it doesn't exist
os.makedirs(output_folder_path, exist_ok=True)

extract_pictures_from_excel(input_excel_path, output_folder_path)
```

```python
import cv2
import pytesseract
import pandas as pd
from PIL import Image

# Set the path to the Tesseract executable
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

# Path to the directory containing the images
images_directory = "C:/Users/Shreshtha/Downloads/Images Directory"

# Function to extract tabular data from an image using OpenCV
def extract_tabular_data(image_path):
    try:
        # Read the image using OpenCV
        img = cv2.imread(image_path)

        # Convert the image to grayscale
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

        # Use a simple thresholding technique to convert to binary image
        _, binary = cv2.threshold(gray, 128, 255, cv2.THRESH_BINARY_INV)

        # Use pytesseract to extract text from the binary image
        text_data = pytesseract.image_to_string(Image.fromarray(binary))

        # Process the extracted text to create a DataFrame
        data = [line.split(':') for line in text_data.split('\n') if ':' in line]
        df = pd.DataFrame(data, columns=["Key", "Value"])
        return df
    except Exception as e:
        print(f"Error extracting tabular data from image {image_path}: {e}")
        return None

# Create an empty DataFrame to store the extracted data
columns = ["Filename", "Tabular Data"]
df = pd.DataFrame(columns=columns)

# Iterate through each image in the directory
for filename in os.listdir(images_directory):
    if filename.endswith(('.png', '.jpg', '.jpeg')):
        image_path = os.path.join(images_directory, filename)
        tabular_data_df = extract_tabular_data(image_path)

        if tabular_data_df is not None:
            # Append the data to the DataFrame
            df = df.append({"Filename": filename, "Tabular Data": tabular_data_df}, ignore_index=True)

# Save the DataFrame to an Excel file
excel_file_path = "extracted_data_with_keys_values.xlsx"
df.to_excel(excel_file_path, index=False)

# Print a summary
print("Data has been saved to:", excel_file_path)
```
