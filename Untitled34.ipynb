```python
import openpyxl
from io import StringIO
import csv

def extract_csv_from_excel(file_path):
    # Load the Excel workbook
    workbook = openpyxl.load_workbook(file_path)

    # Iterate through all sheets in the workbook
    for sheet_name in workbook.sheetnames:
        sheet = workbook[sheet_name]

        # Iterate through all images in the sheet
        for image in sheet._images:
            if image.mime_type == "text/csv":
                print(f"Object in {sheet_name}:")
                print(f"Type: CSV File")
                print(f"Coordinates: {image.anchor}")

                # Extract the CSV data
                csv_data = image.blob
                csv_text = csv_data.decode('utf-8')

                # Use StringIO to create a file-like object for the CSV module
                csv_file = StringIO(csv_text)

                # Parse the CSV data
                csv_reader = csv.reader(csv_file)
                for row in csv_reader:
                    print(row)

                print()

# Example usage
extract_csv_from_excel("C:/Users/Shreshtha/Downloads/Sample MNS Bartrack Report1.xlsx")
```

```python
import openpyxl
from io import StringIO
import csv
import os

def extract_csv_from_excel(file_path):
    # Load the Excel workbook
    workbook = openpyxl.load_workbook(file_path)

    # Iterate through all sheets in the workbook
    for sheet_name in workbook.sheetnames:
        sheet = workbook[sheet_name]

        # Iterate through all images in the sheet
        for image in sheet._images:
            file_extension = os.path.splitext(image.filename)[1]

            if file_extension.lower() == ".csv":
                print(f"Object in {sheet_name}:")
                print(f"Type: CSV File")
                print(f"Coordinates: {image.anchor}")

                # Extract the CSV data
                csv_data = image.blob
                csv_text = csv_data.decode('utf-8')

                # Use StringIO to create a file-like object for the CSV module
                csv_file = StringIO(csv_text)

                # Parse the CSV data
                csv_reader = csv.reader(csv_file)
                for row in csv_reader:
                    print(row)

                print()

# Example usage
extract_csv_from_excel("C:/Users/Shreshtha/Downloads/Sample MNS Bartrack Report1.xlsx")
```

```python
import openpyxl
from io import BytesIO
import csv

def extract_csv_from_excel(file_path):
    # Load the Excel workbook
    workbook = openpyxl.load_workbook(file_path)

    # Iterate through all sheets in the workbook
    for sheet_name in workbook.sheetnames:
        sheet = workbook[sheet_name]

        # Iterate through all images in the sheet
        for image in sheet._images:
            print(f"Object in {sheet_name}:")
            print(f"Type: Unknown")

            # Extract the object data
            object_data = image.blob

            # Attempt to parse the data as CSV
            try:
                csv_text = object_data.decode('utf-8')
                csv_file = BytesIO(csv_text)
                csv_reader = csv.reader(csv_file)
                
                print("Successfully parsed as CSV:")
                for row in csv_reader:
                    print(row)
            except UnicodeDecodeError:
                print("Unable to parse as CSV. Displaying raw data:")
                print(object_data)

            print()

# Example usage
extract_csv_from_excel("C:/Users/Shreshtha/Downloads/Sample MNS Bartrack Report1.xlsx")
```
