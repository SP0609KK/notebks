```python
country_suicide_rate = df.groupby('country')['suicides_no'].sum().reset_index().sort_values(['suicides_no']).tail(10)

plt.figure(figsize=(12, 6))
plt.bar(country_suicide_rate['country'], country_suicide_rate['suicides_no'])
plt.xticks(rotation=90)
plt.xlabel('Country')
plt.ylabel('Total Suicides')
plt.title('Top 10 Countries by Total Suicides')
plt.tight_layout()

plt.show()
```
